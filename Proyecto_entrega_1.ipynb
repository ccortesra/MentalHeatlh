# **Contexto del problema: Análisis de Sentimientos**

El análisis de sentimiento es una de las aplicaciones de los modelos NLP, esta se basa en identificar la emoción o sentimiento de alguna persona plasmada en un texto, ya sea en una reseña de un producto, un correo electronico, un tweet, un comentario de Instagram, entre otros.

Los beneficiarios potenciales de un proyecto de análisis de sentimientos pueden encontrarse en diversas áreas, por ejemplo:

- **Psicología:** Para ayudar a identificar sintomas de trastornos en una persona.
- **Servicio al cliente:** Para identificar de manera objetiva las opiniones de los usuarios.
- **Política:** Ayuda a los partidos políticos a orientar sus campañas.
